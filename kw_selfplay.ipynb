# notebook for generating kw9x9 selfplay training datas

```python
%cd ~
!wget http://launchpadlibrarian.net/650100598/libssl1.1_1.1.1f-1ubuntu2.17_amd64.deb # Install libssl1.1 dependency
!dpkg -i libssl1.1_1.1.1f-1ubuntu2.17_amd64.deb && rm libssl1.1_1.1.1f-1ubuntu2.17_amd64.deb # Clean up .deb file
!wget https://libzip.org/download/libzip-1.11.2.tar.gz # Install libzip dependency
!tar -zxvf libzip-1.11.2.tar.gz && rm libzip-1.11.2.tar.gz # Clean up .tar.gz file
%cd libzip-1.11.2 # Build and install libzip
!cmake . # Generate Makefiles
!make && make install # Build and install libzip
!ldconfig /usr/local/lib && ldconfig /usr/local/lib64 # Update shared library cache
%cd ..
!rm -rf libzip-1.11.2 # Clean up libzip source files
!git clone https://github.com/lightvector/KataGo.git # Clone KataGo repository
!wget https://github.com/changcheng967/KW9x9/releases/download/katagov1.16.4/katago # Download precompiled KataGo binary
!chmod 777 katago # Make katago executable
!mkdir data && mkdir models # Create data and models directories
%cd models
model_url="https://media.katagotraining.org/uploaded/networks/models_extra/kata9x9-b18c384nbt-20231025.bin.gz" # Download 9x9 model(can change to other 9x9 models or 19x19 models)
!wget $model_url

# --- Modify the config file ---
!sed -i '16s/dataBoardLen = 19/dataBoardLen = 9/' /root/KataGo/cpp/configs/training/selfplay1_maxsize9.cfg # Change board size to 9(can be removed if training on 19x19)
```

```python
rm -rf /root/data/* # Clear previous data if any
```

```python
max_games_total = 5000
config_file_path = "./KataGo/cpp/configs/training/selfplay1_maxsize9.cfg" # katago official selfplay config file for 9x9(can be changed to other config files under the same directory)
game_threads = 32
%cd ~
!./katago selfplay -models-dir models -config {config_file_path} -output-dir data -max-games-total {max_games_total} -override-config numGameThreads={game_threads}
!zip -r data.zip data
```
